### Exploratory Data Analysis (EDA)

Steps covered:
- Data loading and inspection
- Summary statistics
- Missing value analysis
- Univariate and bivariate visualizations
- Insights and observations

```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load the dataset
df = pd.read_csv('dataset/Salary_Data.csv')
df.head()
```

```python
# Data overview
print('Shape:', df.shape)
df.info()
df.describe(include='all')
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Univariate Analysis
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.histplot(df['YearsExperience'], bins=10, kde=True, color='skyblue')
plt.title('Years of Experience Distribution')
plt.subplot(1,2,2)
sns.histplot(df['Salary'], bins=10, kde=True, color='salmon')
plt.title('Salary Distribution')
plt.tight_layout()
plt.show()

# Boxplots
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.boxplot(y=df['YearsExperience'], color='lightgreen')
plt.title('Years of Experience Boxplot')
plt.subplot(1,2,2)
sns.boxplot(y=df['Salary'], color='lightcoral')
plt.title('Salary Boxplot')
plt.tight_layout()
plt.show()
```

```python
# Bivariate Analysis
plt.figure(figsize=(7,5))
sns.scatterplot(x='YearsExperience', y='Salary', data=df, color='purple')
plt.title('Years of Experience vs Salary')
plt.show()

# Correlation Heatmap
plt.figure(figsize=(4,3))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
