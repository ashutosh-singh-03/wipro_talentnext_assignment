### Exploratory Data Analysis (EDA)

Steps covered:
- Data loading and inspection
- Summary statistics
- Missing value analysis
- Univariate and bivariate visualizations
- Insights and observations

```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load the dataset
df = pd.read_csv('dataset/Social_Network_Ads.csv')
df.head()
```

```python
# Data overview
print('Shape:', df.shape)
df.info()
df.describe(include='all')
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Univariate Analysis
plt.figure(figsize=(16,5))
plt.subplot(1,3,1)
sns.histplot(df['Age'], bins=15, kde=True, color='skyblue')
plt.title('Age Distribution')
plt.subplot(1,3,2)
sns.histplot(df['EstimatedSalary'], bins=15, kde=True, color='salmon')
plt.title('Estimated Salary Distribution')
plt.subplot(1,3,3)
sns.countplot(x='Gender', data=df, palette='Set2')
plt.title('Gender Count')
plt.tight_layout()
plt.show()
```

```python
# Bivariate Analysis
plt.figure(figsize=(7,5))
sns.scatterplot(x='Age', y='EstimatedSalary', hue='Purchased', data=df, palette='Set1')
plt.title('Age vs Estimated Salary by Purchase')
plt.show()

plt.figure(figsize=(7,5))
sns.boxplot(x='Gender', y='EstimatedSalary', data=df, palette='Set2')
plt.title('Estimated Salary by Gender')
plt.show()

plt.figure(figsize=(7,5))
sns.boxplot(x='Purchased', y='Age', data=df, palette='Set3')
plt.title('Age by Purchase')
plt.show()

# Correlation Heatmap
plt.figure(figsize=(6,4))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
